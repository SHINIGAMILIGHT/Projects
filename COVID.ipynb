```python
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim

longitude = []
latitude = []

def findGeocode(country):

    try:
        geolocator = Nominatim(user_agent="your_app_name")
        return geolocator.geocode(country)
    except GeocoderTimedOut:
        return findGeocode(country)

for i in (country_wise_latest["Country/Region"]):

    if findGeocode(i) != None:
        loc = findGeocode(i)
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)

    else:
        latitude.append(np.nan)
        longitude.append(np.nan)
```

```python
country_wise_latest["Longitude"] = longitude
country_wise_latest["Latitude"] = latitude

country_wise_latest
```

```python
fig = px.scatter_mapbox(country_wise_latest, lat="Latitude", lon="Longitude", color="Active", size="Confirmed",
                  mapbox_style="carto-positron",hover_data=['Country/Region'],template="plotly_dark")
fig.show()
```
